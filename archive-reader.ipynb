```python
import os
archiveDir=os.listdir("./archive")
print(archiveDir)
```

```python
matrix="qwertyuiopasdfghjklzxcvbnm1234567890"
import datetime
for folder in archiveDir:
    timestamp=0
    for letter in range(len(folder)):
        timestamp+=matrix.index(folder[letter])*pow(len(matrix), letter)
    
    saveDate=datetime.datetime.fromtimestamp(timestamp/1000)
    print(folder," - ",saveDate)
    
```

```python
matrix="qwertyuiopasdfghjklzxcvbnm1234567890"
import datetime
for folder in archiveDir:
    timestamp=0
    for letter in range(len(folder)):
        timestamp+=matrix.index(folder[letter])*pow(len(matrix), letter)
    
    saveDate=datetime.datetime.fromtimestamp(timestamp/1000)
    print(folder," - ",saveDate)
    with open("./archive/"+folder+"/_DEBUGRAPORT_") as dr:
        for line in dr:
            print("\t",line.strip())
```

```python
import json
for folder in archiveDir:
    print("\n\n\t\t",folder,"\n")
    listSession=os.listdir("./archive/"+folder)
    for raport in listSession:
        if raport!="_DEBUGRAPORT_":
            print("\n\t",raport,"\n")
            with open("./archive/"+folder+"/"+raport) as dr:
                print(json.dumps(json.loads(dr.read()), sort_keys=True, indent=4))
```
